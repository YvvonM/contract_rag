```python
from pinecone import Pinecone

# initialize connection to pinecone
api_key = "22d53550-c189-4731-8e46-d121abc42e7f"

# configure client
pc = Pinecone(api_key=api_key)
```

```python
#creating an index
import time
from pinecone import ServerlessSpec

index_name = 'contract'
existing_indexes = [
    index_info["name"] for index_info in pc.list_indexes()
]

# creating index
if index_name not in existing_indexes:
    # if does not exist, create index
    pc.create_index(
        index_name,
        dimension=shape[1],  # dimensionality of ada 002
        metric='cosine',
        spec=ServerlessSpec(
        cloud="aws",
        region="us-west-2"
    )

    )
    # wait for index to be initialized
    while not pc.describe_index(index_name).status['ready']:
        time.sleep(1)

# connect to index
index = pc.Index(index_name)
time.sleep(1)
# view index stats
index.describe_index_stats()
```

```python
batch_size = 128

ids = [str(i) for i in range(shape[0])]
# create list of metadata dictionaries
meta = [{'text': chunk} for chunk in chunks]

# create list of (id, vector, metadata) tuples to be upserted
to_upsert = list(zip(ids, embeds, meta))

for i in range(0, shape[0], batch_size):
    i_end = min(i+batch_size, shape[0])
    index.upsert(vectors=to_upsert[i:i_end])

# let's view the index statistics
print(index.describe_index_stats())
```

```python
query = "How much is the escrow amount?"

# create the query embedding
xq = co.embed(
    texts=[query],
    model='embed-english-v3.0',
    input_type='search_query',
    truncate='END'
).embeddings

print(np.array(xq).shape)
```

```python
# Use keyword arguments for the query method
result = index.query(index_name='contract', vector=xq, top_k=5, include_metadata=True)
```
